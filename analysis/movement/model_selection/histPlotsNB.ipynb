```python
from matplotlib import pylab as plt
import matplotlib
import numpy as np
from scipy.stats import norm
#%matplotlib notebook
%matplotlib inline
matplotlib.rcParams.update({'font.size': 16})
```

```python
rho_m = np.load('mc_data/crw_rho_m.npy')

plt.hist(rho_m,normed=True,range=[0.930,0.945],bins=60,alpha=0.8)

plt.xlim(0.930,0.945)
plt.xlabel(r'$\rho$')
plt.ylabel('posterior density')
plt.xticks(np.arange(0.930, 0.946,0.005))
plt.tight_layout()
plt.savefig('crw.png',dpi=300)
plt.show()
print('rho_m mean: ' + str(np.mean(rho_m)) + ' sd: ' + str(np.std(rho_m)))
print('rho_m 95% CI : ' + str(np.percentile(rho_m,2.5)) + ',' + str(np.percentile(rho_m,97.5)))
```

```python
rho_s=np.load('mc_data/cma_rho_s.npy')
alpha=np.load('mc_data/cma_alpha.npy')
beta=np.load('mc_data/cma_beta.npy')

ia = np.load('mc_data/cma_at_a.npy')
il = np.load('mc_data/cma_at_l.npy')
aw = np.load('mc_data/cma_aw.npy')
ig = np.load('mc_data/cma_ig.npy')

allData = [rho_s,alpha,ia,il,aw,ig]

for i in range(len(allData)):
    
    pltThis = allData[i]
    xstart = np.mean(pltThis)-5*np.std(pltThis)
    xstop = np.mean(pltThis)+5*np.std(pltThis)
    X_plot = np.linspace(xstart, xstop, 1000)#[:, np.newaxis]
    log_dens = norm.pdf(X_plot,np.mean(pltThis),np.std(pltThis))
    plt.plot(X_plot, (log_dens),'k-',linewidth=2)
    plt.hist(pltThis,normed=True,range=[xstart,xstop],bins=50,alpha=0.4)
    plt.fill_between(X_plot,np.zeros_like(X_plot), log_dens,alpha=0.4)#, fc='#AAAAFF')

    plt.xlim(xstart,xstop)
    plt.show()
    plt.figure()

print('rho_s ' + str(np.mean(rho_s)))
print('alpha ' + str(np.mean(alpha)))
print('beta ' + str(np.mean(beta)))
print('ia ' + str(np.mean(ia)))
print('il ' + str(np.mean(il)))
print('aw ' + str(np.mean(aw)))
print('ig ' + str(np.mean(ig)))
```
